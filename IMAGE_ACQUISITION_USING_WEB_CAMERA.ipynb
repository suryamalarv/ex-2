```python
#import libraries
import cv2
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time
```

```python
#on the webcamera and read the image
capture=cv2.VideoCapture(0)
ret, frame=capture.read()
```

```python
#save as frame and stop the webcamera
if ret:
    cv2.imwrite("captured.jpg",frame)
capture.release()
```

```python
#read the image to display
read_image=cv2.imread("captured.jpg")
```

```python
#Display the captured image
plt.imshow(read_image[:,:,::-1])
plt.title("Captured Image")
plt.axis('off')
plt.show()
```

```python
#video capture
captured=cv2.VideoCapture(0)
for i in range(50):
    ret, frame =captured.read()
    if not ret:
        break
    frame_rgb_color=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB) #convert to BGR to RGB
    clear_output(wait=True) #wait for clear image
    plt.imshow(frame_rgb_color)
    plt.axis('off')
    plt.show()
    time.sleep(0.05)
captured.release()
```

```python
#Resizing the window
captured=cv2.VideoCapture(0)
for i in range(50):
    ret, frame=captured.read()
    if not ret:
        break
    resized_frame=cv2.resize(frame,(50,100))
    frame_rbg_color=cv2.cvtColor(resized_frame,cv2.COLOR_BGR2RGB)
    clear_output(wait=True)
    plt.imshow(frame_rgb_color)
    plt.title('Resizing the window')
    plt.axis('off')
    plt.show()
    time.sleep(0.05)
captured.release()
```

```python
#Rotate and display the image
rotate = cv2.VideoCapture(0)
for i in range(50):
    ret, frame = rotate.read()
    if not ret:
        break
    rotated_frame = cv2.rotate(frame, cv2.ROTATE_90_CLOCKWISE)
    frame_rgb_color = cv2.cvtColor(rotated_frame, cv2.COLOR_BGR2RGB)
    clear_output(wait=True)
    plt.imshow(frame_rgb_color)
    plt.title('Rotate video')
    plt.axis('off')
    plt.show()
    time.sleep(0.07)
rotate.release()
```
